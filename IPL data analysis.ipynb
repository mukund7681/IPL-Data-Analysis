```python
#extracting record where team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
batting_first
```

```python
batting_first.value_counts()
```

```python
#making a histogram 
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title('distrubution of  runs')
plt.xlabel('runs')
plt.show()
```

```python
#finding out the no of team w.r.t each team won after batting first
batting_first['winner'].value_counts()
```

```python
# making a bar plot for top 3 teams with most win after batting first
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),batting_first['winner'].value_counts()[0:3],color=['blue','yellow','red'])
plt.show()
```

```python
#making pie char 
plt.figure(figsize=(6,6))
plt.pie(list(batting_first['winner'].value_counts()),labels = list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extracting record where team won bowling first

bowling_first=ipl[ipl['win_by_wickets']!=0]
bowling_first
```

```python
#making a histogram 
plt.figure(figsize=(5,4))
plt.hist(bowling_first['win_by_wickets'],bins=30)
plt.title('distrubution of  runs')
plt.xlabel('runs')
plt.show()
```

```python
#finding out the no of team w.r.t each team won after bowling first
bowling_first['winner'].value_counts()
```

```python
# making a bar plot for top 3 teams with most win after bowling first
plt.figure(figsize=(6,6))
plt.bar(list(bowling_first['winner'].value_counts()[0:3].keys()),bowling_first['winner'].value_counts()[0:3],color=['black','blue','red'])
plt.show()
```

```python
#making pie char 
plt.figure(figsize=(6,10))
plt.pie(list(bowling_first['winner'].value_counts()),labels = list(bowling_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
